# Logical Operators
- 
- Logical operator are useful when checking a condition is true or not.
- Python has three logical operators: and,not,or.
- All logical operators returns a boolean value True or False depending on the condition in which it is used.
- We can make condition from comparision operator and check then either they are True or false by the help of logical operators.
- and: true if both the operands are True.
- or: true if either of the operand is true.
- not: true if the operand is false.

```python
print(3 and 4)
print(8 or 9)
a="hii"
b="world"
print(a and b)
print(a or b)
x=9
y=8
print(x>10 and y>6)
print(x<10 and y>6)
print(x<10 or y<6)
print(3 and not 4)
print(3 or not 4)
```

## Precedence of Logical Operators
- When you mix the logical operators in an expression, Python will evaluate them in the order which is called the operator precedence.   
The following shows the precedence of the not, and, and or operator
- 

	- Low
Based on these precedences, Python will group the operands for the operator with the highest precedence first, then group the operf     or       the operator with the lower precedence, nd s-o on.

In case an expression has several logical operators with the same precedence, Python will evaluate them from the left to right:

a or b and c	means	a or (b and c)
a and b or c and d	means	(a and b) or (c and d)
a and b and c or d	means	((a and b) and c) or d
not a and b or c	means	((not a) and b) or c
