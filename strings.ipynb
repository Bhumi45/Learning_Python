# What is indexing

- Indexing means accessming elements individual characters
- There are two types of indexing : positive and negative
- Positive indexing starts from 0 or starts from left side
- Space is also a character
- Negative indexing starts from right-hand side and starts with -1

```python
# positive indexing
a="hello world"
a[2]
```

```python
#negative indexing
a="hello world"
a[-3]
```

# Slicing

- Slicing means accessing a subset of a string based on its indices.
- Again there are two types of slicing: Positive and Negative.
- Denoted by [:]
- Starting point by default in slicing is starting index.
- Ending point by default in slicing in infinity.
- In the range the index after the colon is not included.

```python
a="hello world"
a[:]
```

```python
a="beautiful world"  # includes 1 and do not include 5
a[1:5]
```

```python
a="beautiful world"
a[0:10]
```

```python
a="hello world"
a[-12:0] 
```

```python
a="hello world"
a[0:-1]
```
